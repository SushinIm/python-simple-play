```python
import os
import json
import xmltodict
from haversine import haversine
```

```python
data_path = '100대명산/'
output_path = 'export/'
mntn_list = os.listdir(data_path)
```

#### 테이블 두개 합친 export

```python
mountain = 1
addtional_list = []
for mntn in mntn_list:
    course_list = os.listdir(data_path + mntn + '/')
    course = 1
    for crs in course_list:
        adding = {}
        adding['routeid'] = 'RT' + str(mountain).zfill(3) + str(course).zfill(3)
        adding['routename'] = mntn + '_' + str(course).zfill(3)
        adding['mntnid'] = 'MT' + str(mountain).zfill(3)
        addtional_list.append(adding)

        course += 1
    mountain += 1
```

```python
# 등산로 매핑 테이블
with open(output_path + 'MRMAPPING.json', mode='w', encoding='utf-8') as rt:
    json.dump(addtional_list, rt, indent=4, ensure_ascii=False)
```

#### 등산로 거리 추가 파일 export
- haversine 사용 위도 경도 튜블 두개, 단위 입력

```python
mountain = 1
addtional_list = []
for mntn in mntn_list:
    course_list = os.listdir(data_path + mntn + '/')
    course = 1
    for crs in course_list:
        adding = {}
        with open(data_path + mntn + '/' + crs, mode='r', encoding='utf-8') as f:
            data = json.loads(json.dumps(xmltodict.parse(f.read())))['gpx']

            routes_temp = data['trk']['trkseg']
            routes = []
            if type(routes_temp) == type([]):
                for _ in routes_temp:
                    routes += _['trkpt']
            else:
                routes = routes_temp['trkpt']
                
            adding['routeid'] = 'RT' + str(mountain).zfill(3) + str(course).zfill(3)
            adding['routename'] = mntn + '_' + str(course).zfill(3)
            adding['mntnid'] = 'MT' + str(mountain).zfill(3)

            before = (float(routes[0]['@lat']), float(routes[0]['@lon']))
            distance = 0
            for route in routes:
                compare = (float(route['@lat']), float(route['@lon']))
                distance += haversine(before, compare, unit='m')
                before = compare

            adding['distance'] = round(distance/1000, 1)
        addtional_list.append(adding)
        course += 1
    mountain += 1
```
